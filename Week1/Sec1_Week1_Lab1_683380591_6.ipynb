```python
age = 19
GPA = 3.79
My_favorite_color = "purple"
student = True
print("My age : ",age," ",type(age))
print("My GPA ; ",GPA," ",type(GPA))
print("My favorite color : ",My_favorite_color," ",type(My_favorite_color))
print("Student : ",student," ",type(student))
```

```python
# Puzzler 1
x = 10
y = 3.5
z = x + y
print(z)
print(type(z))
```

ได้ผลลัพธ์: 13.5 มี type เป็น float

เพราะ x มี type เป็น integer และ y มี type เป็น float เมื่อนำมาบวกกันจะได้เป็น float เนื่องจาก float บวกกับ integer จะได้ float เสมอ

```python
# Puzzler 2
name = "Alice"
age = "30" # Note: This is a string
message = name + " is " + age + " years old."
print(message)
print(type(age))
```

ได้ผลลัพธ์: Alice is 30 years old. มี type เป็น string

เพราะ name มี type เป็น string และ age มี type เป็น string เมื่อนำ string มาบวกกันจะเป็นการต่อข้อความกันแทน ทำให้ message เป็น string ที่เป็นข้อความที่ต่อกัน

```python
# Puzzler 3
is_sunny = True
is_raining = False
print(is_sunny and is_raining)
print(type(is_sunny))
```

ได้ผลลัพธ์: False มี type เป็น boolean
