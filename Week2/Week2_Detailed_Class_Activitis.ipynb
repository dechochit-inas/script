### ⏱️ นาทีที่ 4: ตัวแปร (Variables) และชนิดข้อมูลพื้นฐาน

ตัวแปรเปรียบเสมือนป้ายชื่อที่ใช้ชี้ไปยังข้อมูลที่ถูกเก็บอยู่ในหน่วยความจำ การตั้งชื่อนิยมใช้รูปแบบ `snake_case` (อักษรพิมพ์เล็กแยกด้วยเครื่องหมายขีดล่าง)

#### ชนิดข้อมูลพื้นฐานที่สำคัญ:

* **Integers (`int`) และ Floats (`float`):** จำนวนเต็ม และจำนวนจริง (ทศนิยม)
* **Booleans (`bool`):** ค่าความจริง มีเพียง `True` หรือ `False`
* **Strings (`str`):** ข้อความหรืออักขระที่ครอบด้วยเครื่องหมายอัญประกาศเดี่ยวหรือคู่ (`'` หรือ `"`)

#### ตัวอย่างโค้ด: การประกาศตัวแปรและการตรวจสอบชนิดข้อมูล

```python
# การประกาศตัวแปรและกำหนดค่า
age = 21                  # int (จำนวนเต็ม)
gpa = 3.85                # float (ทศนิยม)
is_enrolled = True        # bool (ค่าความจริง)
course_name = "Scripting" # str (ข้อความ)

# การตรวจสอบชนิดข้อมูลด้วยฟังก์ชัน type()
print(type(age))          # ผลลัพธ์: <class 'int'>
print(type(gpa))          # ผลลัพธ์: <class 'float'>
print(type(course_name))  # ผลลัพธ์: <class 'str'>

```

### ⏱️ นาทีที่ 5: การรับค่าและการแสดงผล (Input & Output)

เพื่อให้สคริปต์ทำงานได้อย่างยืดหยุ่น เราจำเป็นต้องมีการรับข้อมูลจากผู้ใช้และจัดฟอร์แมตการแสดงผล

* **จุดสำคัญที่ต้องบันทึก:** ฟังก์ชัน `input()` **จะส่งคืนค่ากลับมาเป็นข้อความ (`str`) เสมอ** หากต้องการนำไปคำนวณทางคณิตศาสตร์ จะต้องนำมาแปลงชนิดข้อมูล (Casting) เสียก่อน ด้วยฟังก์ชัน `int()` หรือ `float()`

#### ตัวอย่างโค้ด: โปรแกรมรับค่าและแปลงข้อมูล (พร้อมแนวคิดพื้นฐานของ Learning Logging)

```python
# 1. การรับข้อมูลประเภทข้อความ
user_name = input("กรุณากรอกชื่อของคุณ: ")

# 2. การรับข้อมูลตัวเลขและทำการแปลงประเภท (Type Casting)
# เนื่องจาก input() ได้ค่าเป็นข้อความ จึงต้องแปลงเป็น float เพื่อใช้คำนวณ
raw_hours = input("สัปดาห์นี้คุณใช้เวลาเรียน Python ไปกี่ชั่วโมง? ")
hours_studying = float(raw_hours)  

# 3. การคำนวณทางคณิตศาสตร์
daily_average = hours_studying / 7

# 4. การแสดงผลข้อมูลออกทางหน้าจอ (Console Output)
print(f"สวัสดีคุณ {user_name}!")
print(f"เฉลี่ยแล้วคุณเรียนวันละ: {daily_average:.2f} ชั่วโมง")

```

### ⏱️ นาทีที่ 1: แนวคิดเรื่อง Control Flow และตัวดำเนินการเปรียบเทียบ

**Control Flow (การควบคุมลำดับการทำงาน)** คือการกำหนดทิศทางให้โปรแกรมตัดสินใจเลือกทำงานตามเงื่อนไขที่กำหนด โดยใช้ **Relational Operators** (ตัวดำเนินการเปรียบเทียบ) เพื่อคืนค่าเป็น Boolean (`True` หรือ `False`) เช่น `==`, `!=`, `<`, `>`, `<=`, `>=`

นอกจากนี้ยังใช้ **Logical Operators** (ตัวดำเนินการทางตรรกศาสตร์) ในการเชื่อมเงื่อนไข:

* `and`: ต้อง True ทั้งคู่
* `or`: True อย่างน้อยหนึ่งอย่าง
* `not`: กลับค่าจาก True เป็น False หรือ False เป็น True

---

### ⏱️ นาทีที่ 2: โครงสร้างเงื่อนไข `if-elif-else`

Python ใช้คำสั่ง `if`, `elif` (else if), และ `else` ในการจัดการสถานการณ์แยกย่อย โดยมีจุดเด่นหลักคือ **การใช้ Indentation (การเยื้องโค้ด)** ในการแบ่งบล็อกการทำงาน

#### ตัวอย่างโค้ด: การตัดเกรดและการตรวจสอบเงื่อนไขซ้อนกัน (Nested Conditionals)

```python
# รับค่าคะแนนจากผู้ใช้และแปลงเป็น float
score = float(input("กรุณากรอกคะแนน (0-100): "))

# ตรวจสอบเงื่อนไขแบบหลายทางเลือก (if-elif-else)
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"ผลการเรียนของคุณคือ: {grade}")

```

---

### ⏱️ นาทีที่ 3: List Comprehension คืออะไร?

**List Comprehension** คือโครงสร้างไวยากรณ์ของ Python ที่ช่วยให้เราสามารถสร้าง List ใหม่จาก Iterable (เช่น list, tuple, range) ได้อย่างกระชับและรวดเร็วในบรรทัดเดียว แทนที่จะต้องเขียนลูป `for` แบบปกติหลายบรรทัด อีกทั้งยังทำงานได้เร็วขึ้น (Speed Increase)

**รูปแบบพื้นฐาน:**
`[ รูปแบบผลลัพธ์ for สมาชิกแต่ละตัว in ข้อมูลต้นทาง ]`

**รูปแบบที่มีการใส่เงื่อนไข (Conditional):**
`[ รูปแบบผลลัพธ์ for สมาชิกแต่ละตัว in ข้อมูลต้นทาง if เงื่อนไข ]`

---

### ⏱️ นาทีที่ 4: ตัวอย่างโค้ด List Comprehension ที่สำคัญ

เราสามารถประยุกต์ใช้การคำนวณและการคัดกรองข้อมูล (Filtering) ร่วมกับ List Comprehension ได้ดังนี้:

```python
# 1. หาค่ายกกำลังสองของเลข 0 ถึง 9
squares = [x**2 for x in range(10)]
print("Squares:", squares)
# ผลลัพธ์: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# 2. คัดเลือกเฉพาะเลขที่หารด้วย 5 ลงตัว และน้อยกว่า 100 (Filtering ด้วย if)
multiples_of_5 = [x for x in range(100) if x % 5 == 0]
print("Multiples of 5:", multiples_of_5)

# 3. การแปลงชนิดข้อมูลใน List (เช่น แปลงข้อความเป็นทศนิยม)
items = ["1.5", "2.3", "3.0", "4.2"]
floats = [float(i) for i in items]
print("Converted Floats:", floats) # ผลลัพธ์: [1.5, 2.3, 3.0, 4.2]

```

### ⏱️ นาทีที่ 5: การประยุกต์ใช้ร่วมกับการทำ Learning Log

เมื่อนำทั้ง 2 เรื่องมารวมกัน เราสามารถสร้างสคริปต์ขนาดเล็กเพื่อจำลองระบบคัดกรอง Log หรือตรวจสอบข้อมูลพฤติกรรมการเรียนเพื่อบันทึกสถิติได้:

```python
# ข้อมูลจำลอง: เวลาที่ใช้ทบทวนบทเรียนในแต่ละวัน (หน่วย: นาที)
study_logs = [45, 120, 30, 0, 150, 90, 15]

# ใช้ List Comprehension คัดกรองเฉพาะ "วันที่ตั้งใจเรียนเป็นพิเศษ" (เรียนมากกว่า 60 นาที)
productive_days = [time for time in study_logs if time > 60]

print(f"สัปดาห์นี้มีวันที่มีประสิทธิภาพสูงทั้งหมด {len(productive_days)} วัน")
print(f"รายการเวลาที่ใช้ในวันเหล่านั้น: {productive_days} นาที")

# ใช้ if-else แสดงข้อความสรุปตามจำนวนวันที่มีประสิทธิภาพ
if len(productive_days) >= 4:
    print("สถานะสัปดาห์นี้: ยอดเยี่ยมมาก! รักษามาตรฐานไว้")
else:
    print("สถานะสัปดาห์นี้: ลองหาเวลาทบทวนเพิ่มขึ้นในสัปดาห์หน้า")

```

<br>*(Relational & Logical Operators)* | เลือกใช้ตัวดำเนินการเปรียบเทียบ (`==`, `>`, `!=`) และตรรกศาสตร์ (`and`, `or`, `not`) ร่วมกันได้อย่างแม่นยำ เพื่อลดความซับซ้อนของเงื่อนไข | ใช้ตัวดำเนินการเปรียบเทียบและตรรกศาสตร์ได้ถูกต้องตามจุดประสงค์ แต่อาจเขียนเงื่อนไขยาวเกินไป แทนที่จะรวบคำสั่งให้กระชับ | ใช้ตัวดำเนินการผิดประเภทในบางจุด (เช่น ใช้ `and` แทนที่จะเป็น `or`) ทำให้ผลลัพธ์ตรรกะคลาดเคลื่อนในบางกรณี | ไม่เข้าใจการทำงานของตัวดำเนินการ ส่งผลให้การเปรียบเทียบหรือการเชื่อมตรรกะผิดพลาดทั้งหมด |
| **3. การคิดเชิงตรรกะและการควบคุมทิศทาง** <br>

<br>*(Logical Thinking & Program Flow)* | โปรแกรมเปลี่ยนทิศทางการทำงานตามข้อมูลที่รับเข้าได้อย่างถูกต้องแม่นยำ 100% แสดงถึงการคิดวิเคราะห์เชิงตรรกะที่เป็นระบบ | โปรแกรมทำงานได้ถูกต้องตามลำดับขั้นตอนส่วนใหญ่ ทิศทางการไหลของข้อมูล (Flow) มีความลื่นไหลและเข้าใจง่าย | โครงสร้างการควบคุมสับสนเล็กน้อย มีบล็อกโค้ดบางส่วนที่ไม่ถูกเรียกใช้งานเลย (Dead Code) เนื่องจากการวางตรรกะผิดพลาด | ทิศทางการทำงานของโปรแกรมไม่เป็นไปตามตรรกะที่ควรจะเป็น โปรแกรมข้ามขั้นตอนหรือทำงานไม่ตรงตามโจทย์ |
| **4. การจัดการกรณีขอบเขตและข้อผิดพลาด** <br>

<br>*(Handling Edge Cases Gracefully)* | มีการดักจับข้อมูลที่อยู่นอกเหนือความคาดหมาย (Edge Cases) ได้ดีเยี่ยม เช่น ค่าติดลบ, ค่าเกิน 100, หรือการพิมพ์พิมพ์เล็ก-ใหญ่ (`.lower()`) | มีการจัดการกับข้อมูลขอบเขตพื้นฐานได้ แต่อาจตกหล่นบางกรณี เช่น ลืมคิดกรณีที่ค่าเท่ากับตัวเลขขอบเขตพอดี (`>=` เทียบกับ `>`) | โปรแกรมทำงานได้เฉพาะกับข้อมูลปกติ (Happy Path) แต่จะค้างหรือให้ผลลัพธ์ผิดพลาดทันทีเมื่อเจอข้อมูลที่แปลกปลอม | ไม่มีระบบตรวจสอบข้อมูลเข้าหรือดักจับ Edge Cases เลย โปรแกรมเกิด Error หรือ Crash ได้ง่ายเมื่อข้อมูลไม่สมบูรณ์ |

### ⏱️ [0-5 นาที] ช่วงที่ 1: แนวคิดของ Control Flow และทิศทางโปรแกรม

โดยปกติแล้ว โปรแกรมของ Python จะรันจากบนลงล่างแบบเส้นตรง (Sequential) แต่ในชีวิตจริง เราจำเป็นต้องให้โปรแกรมตัดสินใจเลือกทำทางใดทางหนึ่งตามสถานการณ์ **Control Flow (การควบคุมทิศทาง)** จึงเปรียบเสมือน "ทางแยกบนถนน" ที่จะเปิดให้สคริปต์ตัดสินใจว่าจะเลี้ยวไปทางไหนตามเงื่อนไขที่ได้รับ

### ⏱️ [5-12 นาที] ช่วงที่ 2: ตัวดำเนินการเปรียบเทียบและตรรกศาสตร์ (Operators)

ก่อนที่โปรแกรมจะเลือกทางเดินได้ มันต้องทำการ "ประเมินค่า" ก่อน โดยผลลัพธ์ของการประเมินจะต้องออกมาเป็นค่าความจริง (**Boolean: `True` หรือ `False**`) เสมอ

#### 1. Relational Operators (ตัวดำเนินการเปรียบเทียบ)

ใช้เปรียบเทียบค่าสองค่า ไม่ว่าจะเป็นตัวเลขหรือข้อความ (String):

* `==` (เท่ากับ), `!=` (ไม่เท่ากับ)
* `<` (น้อยกว่า), `>` (มากกว่า)
* `<=` (น้อยกว่าหรือเท่ากับ), `>=` (มากกว่าหรือเท่ากับ)

#### 2. Logical Operators (ตัวดำเนินการทางตรรกศาสตร์)

ใช้เชื่อมหลายประพจน์หรือปรับเปลี่ยนค่าความจริง:

* `and`: จะให้ค่า `True` **ก็ต่อเมื่อ** เงื่อนไขรอบข้างเป็นจริงทั้งหมด
* `or`: จะให้ค่า `True` ขอแค่มีเงื่อนไขใดเงื่อนไขหนึ่งเป็นจริง
* `not`: ทำการกลับค่าความจริง (Invert) จากจริงเป็นเท็จ และจากเท็จเป็นจริง

#### 💡 ลำดับความสำคัญในการคำนวณ (Order of Operations):

เมื่อเขียนเงื่อนไขรวมกัน Python จะประมวลผลตามลำดับดังนี้:

1. ตัวดำเนินการเปรียบเทียบ (`==`, `>`, `<`, ฯลฯ) จะทำก่อน
2. `not`
3. `and`
4. `or`

```python
# ตัวอย่างการประเมินค่าและเช็กชนิดข้อมูลต่างกัน
print(10 > 5)          # ผลลัพธ์: True
print("apple" == "Apple") # ผลลัพธ์: False (Python ไวต่ออักษรพิมพ์เล็ก-ใหญ่: Case-sensitive)

# การใช้ Logical Operators ร่วมกัน
has_driven_license = True
is_sober = False

# รถจะสตาร์ทติดถ้ามีใบขับขี่ "และ" ไม่เมา
can_drive = has_driven_license and is_sober
print("สามารถขับรถได้:", can_drive)  # ผลลัพธ์: False

```

### ⏱️ [12-22 นาที] ช่วงที่ 3: โครงสร้างการตัดสินใจ (`if`, `if-else`, `if-elif-else`)

ในภาษา Python สิ่งที่สำคัญที่สุดในการกำหนดขอบเขตของเงื่อนไขคือ **เครื่องหมาย Colon (`:`) และการเยื้องโค้ด (Indentation)** โดยมาตรฐานกำหนดให้เคาะช่องว่าง 4 ช่อง

#### 1. คำสั่ง `if` อย่างง่าย

ใช้เมื่อต้องการให้โปรแกรมทำคำสั่งหนึ่งๆ เฉพาะกรณีที่เป็นจริงเท่านั้น ถ้าเป็นเท็จจะข้ามไปเลย

```python
number = 15
if number > 0:
    print(f"{number} เป็นจำนวนบวก") # ทำงานเฉพาะเมื่อเงื่อนไขเป็น True

```

#### 2. คำสั่ง `if-else` (เลือกทำอย่างใดอย่างหนึ่ง)

มีทางเลือกสำรอง (`else`) รองรับกรณีที่เงื่อนไขแรกเป็นเท็จ

```python
number = 7
if number % 2 == 0:
    print(f"{number} เป็นเลขคู่")
else:
    print(f"{number} เป็นเลขคี่") # ทำงานเมื่อเงื่อนไขข้างบนเป็น False

```

#### 3. คำสั่ง `if-elif-else` (หลายทางเลือกที่ขัดแย้งกันเอง - Mutually Exclusive)

**กฎเหล็ก:** โปรแกรมจะไล่ตรวจจากบนลงล่าง และจะ**หยุดทำงานทันที**เมื่อเจอเงื่อนไขแรกที่เป็น `True` ดังนั้น "ลำดับการวางเงื่อนไขจึงสำคัญมาก"

```python
# ดักจับและประมวลผลระบบตัดเกรด (เปรียบเทียบกับการวัดผลในรูบริกส์)
score = float(input("กรุณากรอกคะแนนของนักศึกษา (0-100): "))

# การจัดการ Edge Case พื้นฐาน: ตรวจสอบข้อมูลนำเข้าว่าถูกต้องหรือไม่
if score < 0 or score > 100:
    print("ข้อผิดพลาด: คะแนนต้องอยู่ระหว่าง 0 ถึง 100 เท่านั้น!")
else:
    # เริ่มตรวจเงื่อนไขเรียงลำดับจากมากไปน้อย
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    
    print(f"ระดับคะแนน {score}: นักศึกษาได้เกรด {grade}")

```

### ⏱️ [22-30 นาที] ช่วงที่ 4: การซ้อนเงื่อนไข (Nested Conditionals) และการจัดการ Edge Cases

**Nested Conditionals** คือการใส่บล็อก `if-else` ซ้อนเข้าไปข้างในบล็อก `if-else` อีกชั้นหนึ่ง มักใช้ในกรณีที่ต้องผ่านด่านการคัดกรองชั้นแรกก่อน จึงจะเข้าไปตรวจสอบเงื่อนไขย่อยชั้นที่สองได้

*ข้อควรระวังสำหรับนักศึกษา:* การซ้อนเงื่อนไขมากเกินไปจะทำให้โค้ดอ่านยาก (เรียกว่าเป็นโค้ดแบบ Pyramids of Doom) ควรพยายามใช้ Logical Operators (`and`/`or`) มาช่วยยุบเงื่อนไขหากทำได้

#### โค้ดตัวอย่าง: ระบบคำนวณส่วนลดร้านค้า (ดักจับอักษรพิมพ์เล็ก-ใหญ่เพื่อป้องกันโปรแกรมพัง)

```python
# รับข้อมูลและจัดการ Edge Case เบื้องต้นด้วย .lower() เพื่อเปลี่ยนสตริงเป็นอักษรพิมพ์เล็กทั้งหมด
is_member = input("คุณเป็นสมาชิกของคลับใช่หรือไม่? (yes/no): ").lower().strip()
purchase_amount = float(input("กรุณากรอกยอดซื้อสินค้า ($): "))

# เงื่อนไขชั้นที่ 1: ตรวจสอบสถานะสมาชิก
if is_member == "yes":
    # เงื่อนไขชั้นที่ 2 ย่อย: คำนวณส่วนลดตามยอดซื้อของสมาชิก
    if purchase_amount >= 100:
        discount = purchase_amount * 0.15  # สมาชิกซื้อครบ 100 ได้ลด 15%
        print(f"คุณได้รับส่วนลดสมาชิกขั้นสูง (15%): ${discount:.2f}")
    else:
        discount = purchase_amount * 0.05  # สมาชิกซื้อไม่ถึง 100 ได้ลด 5%
        print(f"คุณได้รับส่วนลดสมาชิกทั่วไป (5%): ${discount:.2f}")

elif is_member == "no":
    # เงื่อนไขชั้นที่ 2 ย่อย: สำหรับบุคคลทั่วไป
    if purchase_amount >= 200:
        discount = purchase_amount * 0.03  # คนทั่วไปต้องซื้อครบ 200 ถึงจะได้ลด 3%
        print(f"คุณได้รับส่วนลดบุคคลทั่วไป (3%): ${discount:.2f}")
    else:
        discount = 0
        print("ยอดซื้อไม่ถึงเกณฑ์ขั้นต่ำ ไม่มีส่วนลดสำหรับบุคคลทั่วไป")

else:
    # ดักจับกรณีผู้ใช้พิมพ์คำอื่นที่นอกเหนือจาก yes/no (Handling edge cases gracefully)
    print("ระบบไม่เข้าใจสถานะสมาชิกของคุณ กรุณากรอกเฉพาะ 'yes' หรือ 'no'")

```
