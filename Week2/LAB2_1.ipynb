#Challenge

```python
num = input("input your number : ")

def check(input_num):

  try:
    num_int = int(input_num)
    print(f"Your type : {type(num_int)}")
    return 0

  except ValueError:
    print(f"Your type : {type(input_num)}")
    print("Your input is not a integer number")
    return 1

def number(number):
  ch = check(number)
  if ch == 0:
    num = int(number)
    if num < 0:
      print("Your number < 0")
    elif num > 0:
      print("Your number > 0")
    else:
      print("Your number = 0")
    print("--------------------------------------------------")
    if (num%2) == 0:
      print("Your number is even number")
    else:
      print("Your number is odd number")

number(num)
```
