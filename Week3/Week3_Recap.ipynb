### แนวคิดสำคัญที่ต้องจดจำ

* **Boolean Expressions (นิพจน์บูลีน):** ประโยคเงื่อนไขที่ให้ผลลัพธ์เป็น `True` (จริง) หรือ `False` (เท็จ) โดยใช้ตัวดำเนินการเปรียบเทียบ เช่น `==`, `!=`, `<`, `>`, `<=`, `>=`
* **Logical Operators (ตัวดำเนินการตรรกศาสตร์):** การเชื่อมเงื่อนไขเข้าด้วยกันโดยใช้ `and` (และ), `or` (หรือ), และ `not` (ไม่)
* **Indentation (การย่อหน้า):** ภาษา Python บังคับให้ใช้การย่อหน้า (แนะนำเป็น 4 Spaces) เพื่อกำหนดว่าโค้ดบรรทัดใดอยู่ในบล็อกเงื่อนไขนั้นๆ

### บล็อกโค้ดที่ 1: การใช้เงื่อนไขพื้นฐานและการตรวจสอบหลายช่วง (if-elif-else)

ตัวอย่างการแบ่งช่วงคะแนนเพื่อตัดเกรด ซึ่งเป็นพื้นฐานของการเช็กเงื่อนไขแบบหลายทางเลือก

```python
# 1. ตัวอย่างคำสั่งเงื่อนไขพื้นฐาน
score = 85

print("--- ตรวจสอบช่วงคะแนนและตัดเกรด ---")
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"คะแนน: {score} | เกรดที่ได้รับ: {grade}")

```

### บล็อกโค้ดที่ 2: การใช้ตัวดำเนินการตรรกศาสตร์เชื่อมหลายเงื่อนไข

ตัวอย่างการผสมคำสั่งตรรกศาสตร์ (`and`, `not`) สำหรับตรวจสอบสถานะความปลอดภัย ระบบจะตรวจสอบว่าต้องตรงตามเงื่อนไขที่กำหนดไว้หรือไม่

```python
# 2. ตัวอย่างการรวมเงื่อนไขด้วยตัวดำเนินการตรรกศาสตร์
is_authenticated = True
has_permission = False
login_attempts = 2
max_attempts = 3

print("\n--- ตรวจสอบสิทธิ์และระบบความปลอดภัย ---")
# ใช้ 'and' และ 'not' ร่วมกันเพื่อเช็กเงื่อนไขที่ซับซ้อนขึ้น
if is_authenticated and has_permission:
    print("ระบบ: อนุญาตให้เข้าถึงข้อมูลได้")
elif is_authenticated and not has_permission:
    print("ระบบ: ปฏิเสธการเข้าถึงเนื่องจากไม่มีสิทธิ์ (No Permission)")
else:
    print("ระบบ: ปฏิเสธการเข้าถึง กรุณาเข้าสู่ระบบก่อน")

# ตรวจสอบสถานะว่าจำนวนครั้งที่ลองเข้าสู่ระบบเกินกำหนดหรือไม่ (เพื่อปูพื้นฐานไปสู่ Loop ในสัปดาห์ถัดไป)
account_locked = login_attempts >= max_attempts
print(f"บัญชีถูกระงับชั่วคราวหรือไม่? {account_locked}")

```
